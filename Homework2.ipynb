```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

### 1. Сгенерировать по 10000 случайных значений из следующих распределений

##### 1.1 Распределение Бернулли с p = 0.35

```python
bernoulli_dist = np.random.binomial(n=1, p=0.35, size=10000)
plt.hist(bernoulli_dist, bins=[-0.5, 0.5, 1.5], edgecolor='black', density=True)
plt.xticks([0, 1])
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Распределение Бернулли (p=0.35)')
plt.show()
```

##### 1.2 Биномиальное распределение с n = 12 и p = 0.2

```python
binomial_dist = np.random.binomial(n=12, p=0.2, size=10000)
plt.hist(binomial_dist, bins=np.arange(0, 13, 1), edgecolor='black', density=True)
plt.xticks(np.arange(0, 13, 1))
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Биномиальное распределение (n=12, p=0.2)')
plt.show()
```

##### 1.3 Гипергеометрическое распределение с n = 7, N = 50, M = 10

```python
hypergeometric_dist = np.random.hypergeometric(ngood=10, nbad=40, nsample=7, size=10000)
plt.hist(hypergeometric_dist, bins=np.arange(-0.5, 8.5, 1), edgecolor='black', density=True)
plt.xticks(np.arange(0, 8, 1))
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Гипергеометрическое распределение (ngood=10, nbad=40, nsample=7)')
plt.show()
```

##### 1.4. Геометрическое распределение с p = 0.7

```python
geometric_dist = np.random.geometric(p=0.7, size=10000)
plt.hist(geometric_dist, bins=np.arange(0.5, 10.5, 1), edgecolor='black', density=True)
plt.xticks(np.arange(1, 11, 1)) 
plt.xlabel('Количество попыток до первого успеха')
plt.ylabel('Частота')
plt.title('Геометрическое распределение (p=0.7)')
plt.xlim(0, 11) 
plt.show()
```

##### 1.5. Распределение Пуассона с λ = 5

```python
poisson_dist = np.random.poisson(lam=5, size=10000)
plt.hist(poisson_dist, bins=np.arange(-0.5, 20.5, 1), edgecolor='black', density=True) 
plt.xticks(np.arange(0, 21, 1)) 
plt.xlabel('Количество событий')
plt.ylabel('Частота')
plt.title('Распределение Пуассона (lambda=5)')
plt.xlim(0, 21) 
plt.show()
```

##### 1.6. Равномерное распределение с a = 5, b = 30

```python
uniform_dist = np.random.uniform(low=5, high=30, size=10000)
plt.hist(uniform_dist, bins=50, edgecolor='black', density=True)
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Равномерное распределение (low=5, high=30)')
plt.show()
```

##### 1.7. Экспоненциальное распределение с λ = 8

```python
exponential_dist = np.random.exponential(scale=8, size=10000)
plt.hist(exponential_dist, bins=50, edgecolor='black', density=True)
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Экспоненциальное распределение (scale=8)')
plt.xlim(0, 50)
plt.show()
```

##### 1.8. Нормальное распределение с a = 2, σ = 5

```python
normal_dist = np.random.normal(loc=2, scale=5, size=10000)
plt.hist(normal_dist, bins=50, edgecolor='black', density=True)
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Нормальное распределение (loc=2, scale=5)')
plt.show()
```

### 2. Для дискретных выборок, полученные в п. 1 построить статистический ряд по значениям (представить в виде dataframe) и полигон частот. Вычислить выборочное среднее и выборочную дисперсию и сравнить их с теоретическими значениями математического ожидания и дисперсии исследуемых распределений.

```python
def analyze_discrete_sample(data, distribution_name, theoretical_mean, theoretical_variance):

    # Статистический ряд
    value_counts = pd.Series(data).value_counts().sort_index()
    statistical_series = pd.DataFrame({'Значение': value_counts.index, 'Частота': value_counts.values})

    # Выборочное среднее и дисперсия
    sample_mean = np.mean(data)
    sample_variance = np.var(data)

    print(f"Анализ {distribution_name}:\n")
    print("Статистический ряд:\n", statistical_series)
    print(f"\nВыборочное среднее: {sample_mean:.4f}, Теоретическое среднее: {theoretical_mean:.4f}")
    print(f"Выборочная дисперсия: {sample_variance:.4f}, Теоретическая дисперсия: {theoretical_variance:.4f}\n")
```

```python
# Анализ распределения Бернулли
analyze_discrete_sample(
    bernoulli_dist,
    "Распределение Бернулли",
    scipy.stats.bernoulli.mean(p=0.35),
    scipy.stats.bernoulli.var(p=0.35)
)
```

```python
# Анализ биномиального распределения
analyze_discrete_sample(
    binomial_dist,
    "Биномиальное распределение",
    scipy.stats.binom.mean(n=12, p=0.2),
    scipy.stats.binom.var(n=12, p=0.2)
)
```

```python
# Анализ гипергеометрического распределения
analyze_discrete_sample(
    hypergeometric_dist,
    "Гипергеометрическое распределение",
    scipy.stats.hypergeom.mean(M=10, n=7, N=50),
    scipy.stats.hypergeom.var(M=10, n=7, N=50)
)
```

```python
# Анализ геометрического распределения
analyze_discrete_sample(
    geometric_dist,
    "Геометрическое распределение",
    scipy.stats.geom.mean(p=0.7),
    scipy.stats.geom.var(p=0.7)
)
```

```python
# Анализ распределения Пуассона
analyze_discrete_sample(
    poisson_dist,
    "Распределение Пуассона",
    scipy.stats.poisson.mean(mu=5),
    scipy.stats.poisson.var(mu=5)
)
```
