### 2. Для дискретных выборок, полученные в п. 1 построить статистический ряд по значениям (представить в виде dataframe) и полигон частот. Вычислить выборочное среднее и выборочную дисперсию и сравнить их с теоретическими значениями математического ожидания и дисперсии исследуемых распределений.

```python
def analyze_discrete_distribution(samples, dist_name, params, theoretical_mean, theoretical_var):
    # Создаем DataFrame с частотами значений
    values, counts = np.unique(samples, return_counts=True)
    df = pd.DataFrame({'Значение': values, 'Частота': counts, 'Относительная частота': counts/len(samples)})
    
    # Вычисляем выборочные характеристики
    sample_mean = np.mean(samples)
    sample_var = np.var(samples, ddof=0)
    
    # Выводим информацию
    print(f"\n{dist_name} распределение с параметрами {params}")
    print(f"Теоретическое среднее: {theoretical_mean:.4f}")
    print(f"Теоретическая дисперсия: {theoretical_var:.4f}")
    print(f"Выборочное среднее: {sample_mean:.4f}")
    print(f"Выборочная дисперсия: {sample_var:.4f}")
    
    # Строим полигон частот
    plt.figure()
    plt.plot(values, counts/len(samples), 'bo-', label='Полигон частот')
    plt.title(f'Полигон частот для {dist_name} распределения')
    plt.xlabel('Значение')
    plt.ylabel('Относительная частота')
    plt.grid(True)
    plt.legend()
    plt.show()
    
    return df
```

```python
np.random.seed(42)
```

```python
p = 0.35
samples = np.random.binomial(1, p, size=10000)
df_bernoulli = analyze_discrete_distribution(
    samples, "Бернулли", f"p={p}", 
    theoretical_mean=p, 
    theoretical_var=p*(1-p)
)
```

```python
n, p = 12, 0.2
samples = np.random.binomial(n, p, size=10000)
df_binomial = analyze_discrete_distribution(
    samples, "Биномиальное", f"n={n}, p={p}", 
    theoretical_mean=n*p, 
    theoretical_var=n*p*(1-p)
)
```

```python
n, N, M = 7, 50, 10
samples = np.random.hypergeometric(M, N-M, n, size=10000)
df_hypergeom = analyze_discrete_distribution(
    samples, "Гипергеометрическое", f"n={n}, N={N}, M={M}", 
    theoretical_mean=n*(M/N), 
    theoretical_var=n*(M/N)*(1-M/N)*((N-n)/(N-1))
)
```

```python
p = 0.7
samples = np.random.geometric(p, size=10000)
df_geometric = analyze_discrete_distribution(
    samples, "Геометрическое", f"p={p}", 
    theoretical_mean=1/p - 1,  # Для определения с 0
    theoretical_var=(1-p)/(p**2)
)
```

```python
lam = 5
samples = np.random.poisson(lam, size=10000)
df_poisson = analyze_discrete_distribution(
    samples, "Пуассона", f"λ={lam}", 
    theoretical_mean=lam, 
    theoretical_var=lam
)
```

### 3. Для непрерывных выборок, полученные в п. 1 построить статистический ряд по интервалам (представить в виде dataframe) и нормированную гистограмму (PDF). Вычислить выборочное среднее и выборочную дисперсию и сравнить их с теоретическими значениями математического ожидания и дисперсии исследуемых распределений.

```python
def analyze_continuous_distribution(samples, dist_name, params, theoretical_mean, theoretical_var, bins=20):
    # Создаем интервалы и считаем частоты
    hist, bin_edges = np.histogram(samples, bins=bins, density=True)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    
    # Создаем DataFrame
    df = pd.DataFrame({
        'Интервал': [f"{bin_edges[i]:.2f}-{bin_edges[i+1]:.2f}" for i in range(len(bin_edges)-1)],
        'Середина интервала': bin_centers,
        'Плотность вероятности': hist
    })
    
    # Вычисляем выборочные характеристики
    sample_mean = np.mean(samples)
    sample_var = np.var(samples, ddof=0)
    
    # Выводим информацию
    print(f"\n{dist_name} распределение с параметрами {params}")
    print(f"Теоретическое среднее: {theoretical_mean:.4f}")
    print(f"Теоретическая дисперсия: {theoretical_var:.4f}")
    print(f"Выборочное среднее: {sample_mean:.4f}")
    print(f"Выборочная дисперсия: {sample_var:.4f}")
    
    # Строим гистограмму
    plt.figure()
    plt.hist(samples, bins=bins, density=True, alpha=0.7, label='Нормированная гистограмма')
    plt.title(f'Нормированная гистограмма для {dist_name} распределения')
    plt.xlabel('Значение')
    plt.ylabel('Плотность вероятности')
    plt.grid(True)
    plt.legend()
    plt.show()
    
    return df
```

```python
a, b = 5, 30
samples = np.random.uniform(a, b, size=10000)
df_uniform = analyze_continuous_distribution(
    samples, "Равномерного", f"a={a}, b={b}", 
    theoretical_mean=(a+b)/2, 
    theoretical_var=(b-a)**2/12
)
```

```python
lam = 8
samples = np.random.exponential(1/lam, size=10000)
df_exponential = analyze_continuous_distribution(
    samples, "Экспоненциального", f"λ={lam}", 
    theoretical_mean=1/lam, 
    theoretical_var=1/(lam**2)
)
```

```python
mu, sigma = 2, 5
samples = np.random.normal(mu, sigma, size=10000)
df_normal = analyze_continuous_distribution(
    samples, "Нормального", f"μ={mu}, σ={sigma}", 
    theoretical_mean=mu, 
    theoretical_var=sigma**2
)
```
